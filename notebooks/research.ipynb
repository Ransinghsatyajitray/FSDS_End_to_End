```python
df
```

```python
(
    ggplot(df, aes(x="variable", y="value", fill="category"))
    + geom_col()
)
```

```python
(
    ggplot(df, aes(x="variable", y="value", fill="category"))
    + geom_col(stat="identity", position="dodge")  # modified
)  
```

```python
dodge_text = position_dodge(width=0.9)  # new

(
    ggplot(df, aes(x="variable", y="value", fill="category"))
    + geom_col(stat="identity", position="dodge", show_legend=False)  # modified
    + geom_text(
        aes(y=-0.5, label="category"),  # new
        position=dodge_text,
        color="gray",
        size=8,
        angle=45,
        va="top",
    )
    + lims(y=(-5, 60))  # new
)
```

```python
dodge_text = position_dodge(width=0.9)

(
    ggplot(df, aes(x="variable", y="value", fill="category"))
    + geom_col(stat="identity", position="dodge", show_legend=False)
    + geom_text(
        aes(y=-0.5, label="category"),
        position=dodge_text,
        color="gray",
        size=8,
        angle=45,
        va="top",
    )
    + geom_text(
        aes(label="value"),  # new
        position=dodge_text,
        size=8,
        va="bottom",
        format_string="{}%",
    )
    + lims(y=(-5, 60))
)
```

```python
from plotnine import *
import numpy as np
import pandas as pd

# Create sample data
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

data = pd.DataFrame({'x': x, 'y': y})

# Create heatmap with binning
plot = (
    ggplot(data, aes(x='x', y='y'))
    + geom_bin2d(bins=30)
    + scale_fill_cmap(name='YlGnBu')  # Yellow-Green-Blue colormap
    + theme_bw()
)
print(plot)
```

```python

# Create sample data
data = pd.DataFrame({
    'category1': ['A', 'A', 'B', 'B', 'C', 'C'],
    'category2': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
    'value': [5, 8, 3, 6, 1, 4]
})

# Create heatmap
plot = (
    ggplot(data, aes(x='category1', y='category2', fill='value'))
    + geom_tile(color='white')
    + scale_fill_gradient(low='white', high='steelblue')
    + theme_classic()
    + coord_fixed()  # Maintain aspect ratio of tiles
)
print(plot)
```

```python
# Add a third category
data['category3'] = ['Group1', 'Group2'] * 3

# Create faceted heatmaps
plot = (
    ggplot(data, aes(x='category1', y='category2', fill='value'))
    + geom_tile(color='white')
    + scale_fill_gradient(low='white', high='steelblue')
    + theme_classic()
    + coord_fixed()
    + facet_wrap('~category3')  # Facet by category3
)
print(plot)
```
